```python
import pandas as pd

filename = 'All metric existing combinations_AUTOCORP_BETA.csv'

# Load the CSV into a DataFrame
df = pd.read_csv(filename, encoding='utf-8')

# Display the first few rows

print(df.head())
```

```python
# help make some unique list for the columns to help filter

# Find the maximum length among unique values across all columns
max_length = max(df.nunique()) + 1   

# Prepare an empty DataFrame with enough rows to accommodate the longest list of unique values
unique_values_df = pd.DataFrame(index=range(max_length))

# Iterate through each column in the original DataFrame
for column in df.columns:
    # Extract unique values, sort them if necessary, and temporarily store them in a list
    unique_values_list = list(df[column].unique())
    
    # If the number of unique values in the current column is less than max_length, extend the list with None values
    unique_values_list.extend([None] * (max_length - len(unique_values_list)))
    
    # Add the extended list of unique values as a new column in the unique_values_df DataFrame
    unique_values_df[column] = unique_values_list


# Save the compiled DataFrame to a single CSV file
unique_values_df.to_csv('compiled_unique_values_helper.csv', index=False, encoding='utf-8')
```

```python
import numpy as np

for column in unique_values_df.columns:
    # Extract unique values without NaNs and convert to a numpy array
    unique_values_array = np.unique(unique_values_df[column].dropna().values)
    
    print(unique_values_array)
```

```python
# 1st level filtering: manual csv control list  
# Import master metric list that is filtered (go into the csvs that was made from prior code and save 
# a new version of only the metrics you want and save it as compiled_unique_values_filtered.csv)

num_rows_original = len(df)
print(num_rows_original)

filtered_values_df = pd.read_csv('compiled_unique_values_filtered.csv')

first_filter_df = df.copy()
# Iterate over each column in the filtered_values_df to filter the original DataFrame
for column in filtered_values_df.columns:
    print('before filter')
    print(df[column].unique())
    # Drop NaN values from the filtered_values_df for the current column to avoid filtering with NaN
    filtered_values = filtered_values_df[column].dropna()
    print(filtered_values)
    
    # Filter the original DataFrame, keeping only rows where the value in the current column matches one of the filtered values
    first_filter_df = first_filter_df[first_filter_df[column].isin(filtered_values)]
    
    print('after filter')
    print(df[column].unique())
    
filtered_df = first_filter_df.copy()
num_rows_post_filter = len(filtered_df)
print(num_rows_post_filter)
```

```python
for column in filtered_df.columns:
    # Extract unique values without NaNs and convert to a numpy array
    unique_values_array = np.unique(filtered_df[column].dropna().values)
    
    print(unique_values_array)
```

```python
# 2nd level filtering: systematic filters and catch-alls (ctrl + / to comment out):
filtered_df = first_filter_df.copy()
print('Before filtering')
print('Initial length:', len(filtered_df))
print(filtered_df['METRIC_NAME'].unique())

# 2. Exclude LIVE in METRIC_NAME or COMPONENT_NAME
filtered_df = filtered_df[~(filtered_df['METRIC_NAME'].str.contains('LIVE', case=False, na=False) |
                            filtered_df['COMPONENT_NAME'].str.contains('LIVE', case=False, na=False))]
print('After removing rows with "LIVE":')
print('Current length:', len(filtered_df))

# 3. Remove Comp JSteps
filtered_df = filtered_df[~((filtered_df['RETAILER_CLASSIFICATION'] == 'Competitor') & 
                            ((filtered_df['METRIC_NAME'] == 'ROC_Step') | (filtered_df['METRIC_NAME'] == 'ROC_Ratio')))]
print('After removing Comp JSteps:')
print('Current length:', len(filtered_df))

# 5. Regional Attributes filtering
filtered_df = filtered_df[~((filtered_df['DEMO_CUT'] != 'Canada') & 
                           ((filtered_df['COMPONENT_NAME'].str.contains('attribute', case=False, na=False)) | (filtered_df['METRIC_NAME'].str.contains('attribute', case=False, na=False))))]
print('After filtering Regional Attributes:')
print('Current length:', len(filtered_df))

# 6. Remove Regional Comp Share
filtered_df = filtered_df[((filtered_df['DEMO_CUT'] == 'Canada') | 
                           (filtered_df['RETAILER_CLASSIFICATION'] == 'Client'))]
print('After removing Regional Comp Share:')
print('Current length:', len(filtered_df))

# 9. Regional only Total company
filtered_df = filtered_df[((filtered_df['DEMO_CUT'] == 'Canada') | 
                           (filtered_df['REPORTING_LEVEL'] != 'Division'))]
print('After applying Regional only Total company filter:')
print('Current length:', len(filtered_df))

# 10. Division no Competitor
filtered_df = filtered_df[((filtered_df['RETAILER_CLASSIFICATION'] == 'Client') | 
                           (filtered_df['REPORTING_LEVEL'] != 'Division'))]
print('After removing Division no Competitor:')
print('Current length:', len(filtered_df))

# 13. No divisional attributes
filtered_df = filtered_df[~((filtered_df['DEMO_CUT'] == 'Division') & 
                            filtered_df['METRIC_NAME'].str.contains('Attribute', case=False, na=False))]
print('After removing divisional attributes:')
print('Current length:', len(filtered_df))

# # 17. No Amazon Brick
# filtered_df = filtered_df[~(filtered_df['RETAILER_NAME'].str.contains('Amazon', case=False, na=False) & 
#                             filtered_df['CHANNEL_CUT'].str.contains('B&M', case=False, na=False))]
# print('After removing Amazon Brick & Mortar:')
# print('Current length:', len(filtered_df))

# 18. No Winners Ecom
filtered_df = filtered_df[~(filtered_df['RETAILER_NAME'].str.contains('Winner', case=False, na=False) & 
                            filtered_df['CHANNEL_CUT'].str.contains('Ecom', case=False, na=False))]
print('After removing Winners Ecom:')
print('Current length:', len(filtered_df))

# 19. No E/E, B/B, O/O attributes
excluded_channel_cuts = ['Omni of Omni', 'B&M of B&M', 'Ecom of Ecom']
filtered_df = filtered_df[~(filtered_df['METRIC_NAME'].str.contains('Attribute', case=False, na=False) & 
                            filtered_df['CHANNEL_CUT'].isin(excluded_channel_cuts))]
print('After removing E/E, B/B, O/O attributes:')
print('Current length:', len(filtered_df))

# 23. No Flyer metrics
filtered_df = filtered_df[~filtered_df['METRIC_NAME'].str.contains('Flyer', case=False, na=False)]
print('After removing Flyer metrics:')
print('Current length:', len(filtered_df))

```

```python
#check CSVs (for 1st or comparison)

# Load the CSV into a DataFrame
file_path = 'PortalControl_2024-04-08.csv'  # Update this to your CSV file path
df1 = pd.read_csv(file_path)


# Create a Pandas Excel writer using XlsxWriter as the engine
excel_file = 'unique_values_with_counts1.xlsx'
excel_writer = pd.ExcelWriter(excel_file, engine='xlsxwriter')

# Create an empty DataFrame for concatenated results
concatenated_results = pd.DataFrame()

# Track the starting column for each new set of results
start_col = 0

# Iterate through each column in the DataFrame
for column in df1.columns:
    # Calculate the count of each unique value in the current column
    value_counts = df1[column].value_counts(dropna=False).reset_index()
    value_counts.columns = [f'{column}', f'{column} Count']  # Rename columns
    
    # If it's the first column, directly assign the value_counts DataFrame
    if concatenated_results.empty:
        concatenated_results = value_counts
    else:
        # For subsequent columns, concatenate horizontally
        concatenated_results = pd.concat([concatenated_results, value_counts], axis=1)
        
    
    # Print the results for the current column
    print(f"Results for column: {column}")
    print(value_counts)
    print("\n")  # Add a newline for readability between columns
        
# Save the concatenated results to an Excel file in one sheet
excel_file = 'unique_values_with_counts_version1.xlsx'
concatenated_results.to_excel(excel_file, index=False, sheet_name='Unique Values & Counts')


excel_writer.close()
```
